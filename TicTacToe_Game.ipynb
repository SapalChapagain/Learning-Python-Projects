```python
from IPython.display import clear_output
import random


#WELCOME TO THE GAME
print('Welcome to Tic Tac Toe!')


#DISPLAY A BOARD 
def display_board(board):
    clear_output()
    print("| "+ board[1]+ " | " + board[2] + " | " + board[3] + " |"  )
    print("| "+ board[4]+ " | " + board[5] + " | " + board[6] + " |"  )
    print("| "+ board[7]+ " | " + board[8] + " | " + board[9] + " |"  )
    
#GET THE PLAYER TO CHOOSE X or O  
def player_input():
    answer = "wrong"
    marker = ""
    while answer not in ["X","O"]:
        answer = input("Choose X or O").upper()
        if answer not in ["X", "O"]:
            print("You HAVE to choose X or O")
            
    if answer == "X":
        firstplayermarker = "X"
        secondplayermarker = "O"
        
    else:
        firstplayermarker == "O"
        secondplayermarker = "X"
        
    return [firstplayermarker, secondplayermarker]
    
        


#PLACE THE MARKER IN THE CORRECT POSITION
def place_marker(board, marker, position):
    board[position] = marker

    
#CHECK IF SOMEONE WON
def win_check(board, mark):
    return ((board[1]== mark and board[2]== mark and board[3] == mark) or     
    (board[4]== mark and board[5]== mark and board[6] == mark) or
    (board[7]== mark and board[8] == mark and board[9] == mark) or 
    (board[1]== mark and board[4]== mark and board[7] == mark) or
    (board[2]== mark and board[5]== mark and board[8] == mark) or 
    (board[3] == mark and board[6]== mark and board[9] == mark) or 
    (board[1] == mark and board[5]== mark and board[9] == mark) or 
    (board[3] == mark and board[5]== mark and board[7] == mark)) 



#CHOOSE WHO GOES FIRST
def choose_first():
    randomnumber = random.randint(1,2)
    firstplayer = ""
    if randomnumber == 1:
        firstplayer = "X"
        secondplayer = "O"
    else:
        firstplayer = "O"
        secondplayer = "X"
    return [firstplayer, secondplayer]


#MAKE SURE THE SPACE WHERE THE PLAYER WANTS TO PUT THEIR MARK IS FREE
def space_check(board, position):
    return board[position] == " "

#CHECK IF ITS A TIE
def full_board_check(board):
    if " " in board:
        return False
    else: 
        return True
    
    
#GET THE PLAYER TO CHOOSE A POSITION
def player_choice(board):
    choice = int(input("Choose a position from 1-9 for your next position"))
    while space_check(board, choice) == False:
        choice = int(input("That space is taken, choose again"))
    return choice
        
#ASK IF THEY WANT TO PLAY AGAIN
def replay():
    answer = input("Do you want to play? Put Y or N")
    
    while answer not in ["Y","N"]:
        answer = input("You have to write Y or N")
    
    return answer == "Y"

        
#GAMEITSELF
while True:
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    firstplayermarker = player_input()[0]
    turn = choose_first()[0]
    print(turn + " will go first")
    play_game = input("Are you ready to play? Enter Yes or No")
    
    if play_game[0].lower() == "y":
        game_on = True
    else:
        game_on = False
    
    while game_on == True:
        if turn == "X":
            print("Player X turn")
            display_board(board)
            position = player_choice(board)
            place_marker(board, "X", position)
            if win_check(board, "X"):
                print("Congrats X won")
                game_on = False
            else:
                if full_board_check(board):
                    display_board(board)
                    print("The game is a tie")
                    break
                else: 
                    turn = "O"
                        
        else:
            print("Player O turn")
            display_board(board)
            position = player_choice(board)
            place_marker(board, "O", position)
            if win_check(board, "O"):
                print("Congrats O won")
                game_on = False
            else:
                if full_board_check(board):
                    display_board(board)
                    print("The game is a tie")
                    break 
                else: 
                    turn = "X"
        
            
                        
    if not replay():
        print("Thanks for playing")
        break
            

    
 



    
```
